```python
# Web scraping fact checking article on Donald Trump statemnts from Politifact article
import requests 
import pandas as pd  
r = requests.get('http://www.politifact.com/personalities/donald-trump/statements/byruling/false/')
from bs4 import BeautifulSoup  
soup = BeautifulSoup(r.text, 'html.parser')  
results = soup.find_all('div', attrs={'class':'scoretable__item'})
#print(len(results))

records = []  
for result in results:  
    date = result.find('span',attrs={'class':'article__meta'}).text
    lie =  result.find('a', attrs={'class':'link'}).text
    fact = result.find('p', attrs={'class':'quote'}).text
    url =  result.find('a')['href'] 
    records.append(( date,lie,fact, url))

df = pd.DataFrame(records, columns=['Date', 'Lie','Fact', 'Url'])  
df
#df.to_csv('trump_lies_wp.csv', index=False, encoding='utf-8')
```

### Scraping the web page into Python
### using Washington Post Pinocchio category articles

```python
import requests  
import pandas as pd  
r = requests.get('https://www.washingtonpost.com/news/fact-checker/wp/category/donald-trump/')

#The code above fetches our web page from the URL, and stores the result in a "response" object called r.
# print the first 500 characters of the HTML
#print(r.text[0:500])
#Parsing the HTML using Beautiful Soup

from bs4 import BeautifulSoup  
soup = BeautifulSoup(r.text, 'html.parser')  
results = soup.find_all('div', attrs={'class':'story-list-item'})

# Lists to store the scraped data in
lies = []
facts = []
author = []
category = []
urls =[]

#headline = [div.text for div in soup.findAll('div', attrs={'class':'story-headline'})]
lie = [h3.text for h3 in soup.findAll('h3')]
fact = [div.text for div in soup.findAll('div', attrs={'class':'story-description'})]
urls = [h3.find('a')['href'] for h3 in soup.findAll('h3')]
author = [span.text for span in soup.findAll('span', attrs={'class':'author'})]
category = [a.text for a in soup.findAll('li', attrs={'class':'category'})]

df = pd.DataFrame(list(zip(lie, fact,urls, author,category)),columns=['Lies','Facts', 'Urls','Author','Category'])
df
#df.to_csv('trump_lies_wp.csv', index=False, encoding='utf-8')   
```
